# Increase intensity

```python
import librosa
import soundfile as sf
import os
import numpy as np

# Define the input and output directories
input_dir = r"C:\Users\44778\OneDrive\Desktop\UWE_Docs\7. Dissertation\CNN Data Files\Cleaned audio\swamphen"
output_dir = r"C:\Users\44778\OneDrive\Desktop\UWE_Docs\7. Dissertation\CNN Data Files\Louder audio\swamphen"

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Set the desired volume increase factor
volume_increase_factor = 10.0  # Increase volume by 100% (i.e., double the amplitude)

# Loop through all files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".wav"):  # Process only .wav files
        input_file = os.path.join(input_dir, filename)
        
        # Load the audio file
        y, sr = librosa.load(input_file, sr=None)
        
        # Increase the volume by scaling the amplitude
        y_louder = y * volume_increase_factor
        
        # Ensure the audio signal is within the valid range [-1.0, 1.0]
        y_louder = np.clip(y_louder, -1.0, 1.0)
        
        # Save the enhanced audio to the output directory
        output_file = os.path.join(output_dir, filename)
        sf.write(output_file, y_louder, sr)

print("Processed and saved all files")
```

```python
import noisereduce as nr
import librosa
import soundfile as sf
import os

# Define the input and output directories
input_dir = r"C:\Users\44778\OneDrive\Desktop\UWE_Docs\7. Dissertation\CNN Data Files\audio file 5 seconds\swamphen"
output_dir = r"C:\Users\44778\OneDrive\Desktop\UWE_Docs\7. Dissertation\CNN Data Files\Cleaned audio\swamphen"

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Loop through all files in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".wav"):  # Process only .wav files
        input_file = os.path.join(input_dir, filename)
        
        # Load the audio file
        y, sr = librosa.load(input_file, sr=None)

        # Perform noise reduction
        y_clean = nr.reduce_noise(y=y, sr=sr)

        # Save the cleaned audio to the output directory
        output_file = os.path.join(output_dir, filename)
        sf.write(output_file, y_clean, sr)

print("Processed and saved all files")
```
