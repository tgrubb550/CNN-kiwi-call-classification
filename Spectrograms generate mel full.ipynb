```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import os
import numpy as np

# Function to plot a spectrogram and save it
def plot_and_save_spectrogram(data, title, ylabel, sr, hop_length, output_file):
    plt.figure(figsize=(12, 6), dpi=150)
    librosa.display.specshow(data, sr=sr, x_axis='time', y_axis='log', hop_length=hop_length, cmap='viridis')
    plt.ylim(256, 4096)  # Set y-axis limits
    plt.axis('off')
    plt.tight_layout()
    plt.savefig(output_file)
    plt.close()

# Set input and output directories - bellbird
input_folder = r"C:\Users\44778\OneDrive\Desktop\UWE_Docs\7. Dissertation\CNN Data Files\audio files (TSM + SSiA)\bellbird"
output_folder = r"C:\Users\44778\OneDrive\Desktop\UWE_Docs\7. Dissertation\CNN Data Files\Old_Data\Viridis\Mel (TSM + SSiA)\bellbird"

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Loop through all files in the input directory
for file_name in os.listdir(input_folder):
    if file_name.endswith(('.wav', '.mp3')):  # Process only audio files
        # Load the audio file
        input_file_path = os.path.join(input_folder, file_name)
        y, sr = librosa.load(input_file_path)

        # Generate Mel Spectrogram
        mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
        mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
        
        # Define the output file path
        output_file = os.path.join(output_folder, f"{os.path.splitext(file_name)[0]}_mel_spec.png")
        
        # Plot and save the Mel Spectrogram
        plot_and_save_spectrogram(mel_spec_db, 'Mel Spectrogram', 'Frequency (Hz)', sr, hop_length=512, output_file=output_file)

print("Processed and saved all Mel Spectrograms")
```
